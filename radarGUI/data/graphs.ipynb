```python
import numpy as np
from math import atan2
from bokeh.io import output_notebook, show
from bokeh.plotting import figure,output_file,show
from bokeh.models import HoverTool
from bokeh.models import ColumnDataSource
output_notebook()
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False 
#https://nbviewer.jupyter.org/github/bokeh/bokeh-notebooks/tree/master/tutorial/
```

```python
#输出文件格式：est_px est_py est_vx est_vy means_px means_py gt_px gt_py gt_vx gt_vy 
state = 0;   #融合数据：state  = 0;  激光雷达数据：state = 1; 毫米波雷达数据：state = 2
file_path = "output.txt" #输出文件所在路径
all_px_est = []
all_py_est = []
all_px_means = []
all_py_means = []
all_px_gt = []
all_py_gt = []



with open(file_path) as f:
    
    for line in f:
        data = line.split() #以空格分割
        est_px = float(data[0])
        est_py = float(data[1])
        est_vx = float(data[2])
        est_vy = float(data[3])
        means_px = float(data[4])
        means_py = float(data[5])
        gt_px = float(data[6])
        gt_py = float(data[7])
        gt_vx = float(data[8])
        gt_vy = float(data[9])
        all_px_est.append(est_px)
        all_py_est.append(est_py)
        all_px_means.append(means_px)
        all_py_means.append(means_py)
        all_px_gt.append(gt_px)
        all_py_gt.append(gt_py)


est_source = ColumnDataSource(data = {
    'x' : all_px_est,
    'y' : all_py_est 
})

means_source = ColumnDataSource(data = {
    'x' : all_px_means,
    'y' : all_py_means 
})

gt_source = ColumnDataSource(data = {
    'x' : all_px_gt,
    'y' : all_py_gt
})
```

```python
if(state == 0):
    p = figure(title = "radar、lidar融合数据曲线",title_location = "above",x_axis_label = 'px',y_axis_label = "py",plot_width = 1000, plot_height = 700)
elif(state == 1):
    p = figure(title = "lidar数据曲线",title_location = "above",x_axis_label = 'px',y_axis_label = "py",plot_width = 1000, plot_height = 700)
else:
    p = figure(title = "radar数据曲线",title_location = "above",x_axis_label = 'px',y_axis_label = "py",plot_width = 1000, plot_height = 700)

#画出估计值
p.line('x', 'y',line_width = 2,line_color = "blue",legend = "扩展卡尔曼估计值",source = est_source)
p.circle('x', 'y', size =3,line_width = 1,fill_color = "blue",line_color = "blue",fill_alpha = 1,legend = "扩展卡尔曼估计值",source = est_source)
#画出测量值
p.circle('x', 'y', size = 4,fill_color = "green",line_color = "green",fill_alpha = 1,line_width = 1,legend = "测量数据",source = means_source)
#画出真实值
p.line('x', 'y',line_color = "red",line_width =2, legend = "真实值",source = gt_source)
p.circle('x', 'y', size =3,line_width = 1, fill_color = "red",line_color = "red",fill_alpha = 1,legend = "真实值",source = gt_source)
p.legend.location = "bottom_right"
show(p)
```

```python
from bokeh.layouts import gridplot



plot_options = dict(width=300, plot_height=300)

# create a new plot
s1 = figure(**plot_options)
s1.line('x', 'y', line_width = 0.5,legend = "扩展卡尔曼估计值", source = est_source)
s1.circle('x', 'y', size =2, fill_color = "blue",line_color = "blue",fill_alpha = 1,legend = "扩展卡尔曼估计值", source = est_source)


# create a new plot and share both ranges
s2 = figure(x_range=s1.x_range, y_range=s1.y_range, **plot_options)
s2.circle('x', 'y', size = 2, fill_color = "green",line_color = "green",fill_alpha = 1,line_width = 1, legend = "测量数据", source = means_source)

# create a new plot and share only one range
s3 = figure(x_range=s1.x_range, **plot_options)
s3.line('x', 'y',line_color = "red",line_width = 0.5, legend = "真实值",source = gt_source)
s3.circle('x', 'y', size =2, fill_color = "red",line_color = "red",fill_alpha = 1,legend = "真实值", source = gt_source)
p = gridplot([[s1, s2, s3]])

# show the results
show(p)
```
